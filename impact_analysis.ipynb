```python
sales_columns_1 = ['Customer Id','PC-SPC','Unit Price','Cost','Sales','Margin']
df_sales_1 = df_sales[sales_columns_1]
df_sales_1.head(5)
```

```python
df_sales_agg1=df_sales_1.groupby(['Customer Id','PC-SPC']).sum().reset_index()
#type(df_sales_agg1)
```

```python
df_sales_agg1['TM%'] = df_sales_agg1['Margin']/df_sales_agg1['Sales']*100
df_sales_agg1[['PC1','PC2']] = df_sales_agg1['PC-SPC'].str.split("-",expand=True,)
```

```python
master_acc = '1470089'
slave_acc = '1461867'
```

```python
df_sales_agg1.head(10)
```

```python
df_sales_agg1['Customer Id'].astype(str)
```

```python
df_sales_agg_master = df_sales_agg1[df_sales_agg1['Customer Id']==master_acc]
#df_sales_agg_master = df_sales_agg_master.rename(columns=lambda s: s+"_mas")
df_sales_agg_master = df_sales_agg_master.add_suffix('_MAS')


df_sales_agg_slave = df_sales_agg1[df_sales_agg1['Customer Id']==slave_acc]
#df_sales_agg_slave = df_sales_agg_slave.rename(columns=lambda s: s+"_sla")
df_sales_agg_slave = df_sales_agg_slave.add_suffix('_SLA')
```

```python
df_sales_agg_master = df_sales_agg_master.rename(columns={'PC-SPC_MAS':'PC-SPC'})
df_sales_agg_slave = df_sales_agg_slave.rename(columns={'PC-SPC_SLA':'PC-SPC'})
```

```python
df_sales_agg_master.head()
```

```python
## Join two data frame with left-join
df_sales_merged = pd.merge(df_sales_agg_slave,df_sales_agg_master,how = 'left',on = 'PC-SPC') 
```

```python
df_sales_merged.head(5)
```

```python
# impact
df_sales_merged['impact'] = df_sales_merged[['TM%_MAS','TM%_SLA','Sales_SLA','Margin_SLA']].apply(lambda x: x['Sales_SLA']*x['TM%_MAS']/100-x['Margin_SLA'] if x['TM%_MAS']<x['TM%_SLA'] else 0,axis=1)
```
